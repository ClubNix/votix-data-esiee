```python
import matplotlib.pyplot as plt
import yaml

file = open('../data/halloffame.yml')
halloffame = yaml.load(file)
archives = halloffame['archives']

def plot_results_pie(archive):
    results = archive['candidates']
    candidates = [c for c in results if c['votes'] > 0]

    labels = [c['name'] + " (" + str(c['votes']) + ")" for c in candidates]
    sizes = [c['votes'] for c in candidates]
    colors = [c['color'] for c in candidates]

    explode = [0.1] + [0] * (len(candidates) - 1)

    fig1, ax1 = plt.subplots()
    fig1.set_size_inches(8.5, 3.5)
    ax1.pie(sizes, explode=explode, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    ax1.axis('equal')

    return plt.show()
```
