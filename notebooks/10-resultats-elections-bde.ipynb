```python
import matplotlib.pyplot as plt
import yaml

file = open('../data/halloffame.yml')
halloffame = yaml.load(file)
archives = halloffame['archives']

def plot_results(results):
    candidates2020 = [c for c in results if c['votes'] > 0]
    labels = [c['name'] + " (" + str(c['votes']) + ")" for c in candidates2020]
    sizes = [c['votes'] for c in candidates2020]

    explode = [0.1] + [0] * (len(candidates2020) - 1)

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    ax1.axis('equal')

    return plt.show()
```
