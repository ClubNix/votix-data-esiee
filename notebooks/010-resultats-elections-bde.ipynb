```python
%matplotlib inline
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import yaml

file = open('../data/halloffame.yml')
halloffame = yaml.load(file, Loader=yaml.FullLoader)
archives = halloffame['archives']

def summarize_candidate(c):
    return c['name'] + " : " + str(c['votes']) + " votes, était " + ('éligible' if c['electable'] else 'non éligible') + " et " + ('votable' if c['votable'] else 'non votable')

def plot_results_pie(archive):
    results = archive['candidates']
    candidates = [c for c in results if c['votable']]

    labels = [c['name'] + " (" + str(c['votes']) + ")" for c in candidates]
    sizes = [c['votes'] for c in candidates]
    colors = [c['color'] for c in candidates]

    explode = [0.1] + [0] * (len(candidates) - 1)

    figs, axs = plt.subplots(1, 2)

    figs.set_size_inches(13, 3)
    axs[0].pie(sizes, explode=explode, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    axs[0].axis('equal')

    axs[1].axis("off")

    clust_data = [(summarize_candidate(c), ) for c in results]
    axs[1].table(cellText=clust_data,loc='center',edges='open')

    plt.show()

def display_year_intro(archive):
    display(HTML('<h2>Edition ' +  str(archive['year']) + ', vainqueur: ' + archive['winner'] + '</h2>'))

def display_year_summary(archive):
    display_year_intro(archive)
    plot_results_pie(archive)
```
